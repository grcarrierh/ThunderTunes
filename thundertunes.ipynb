```python
# imports
import requests
import webbrowser
import spotipy
import sys
import json
import time

# grab weather from wunderground API
def get_wunderground_weather(loc_state, loc_city):
    url = "http://api.wunderground.com/api/2827ad3b88378325/conditions/q/%s/%s.json" % (loc_city, loc_state)
    params = { 'city' : loc_city, 'state' : loc_state }
    response = requests.get(url, params)
    weather_data = response.json()
    tempF = float(weather_data['current_observation']['temp_f'])
    precip = float(weather_data['current_observation']['precip_today_in'])
    conditions = {'tempF' : tempF, 'precip_today': precip}
    print("In %s, %s it is %s degrees right now and has precipitated %.2f inches today.\n" % (loc_state, loc_city, conditions['tempF'], conditions['precip_today']))
    return conditions

# grab track information from spotipy API
def get_track_name(trackcodes):
    for code in trackcodes:
        urn = 'spotify:track:%s' % code
        sp = spotipy.Spotify()
        track_info = sp.track(urn)
        song_name = track_info['name']
        print("Song name:", song_name)
        artist = [artist['name'] for artist in track_info['album']['artists']]
        print("Artist name:", artist[0])
        album_name = track_info['album']['name']
        print("Album name:", album_name)
        print(' ')
    return

# playlist URLs to open Spotify playlists
WarmDry_url = "https://open.spotify.com/user/joelkleiman/playlist/46hFeYgLsDo8bQgpdgSPjw"
WarmPrecip_url = "https://open.spotify.com/user/joelkleiman/playlist/5M7Z9HfneUp2SGBeTVPKPe"
MildDry_url = "https://open.spotify.com/user/joelkleiman/playlist/3lj2QDd0U8N4l8kC8zIDIT"
MildPrecip_url = "https://open.spotify.com/user/joelkleiman/playlist/5k0BRXYYshDoi3x8d8Lh9R"
ColdDry_url = "https://open.spotify.com/user/joelkleiman/playlist/1JjBEymEri6JNrB3m4nfwO"
ColdPrecip_url = "https://open.spotify.com/user/joelkleiman/playlist/1JjBEymEri6JNrB3m4nfwO"

# track URIs to print out track information
WarmDry_trackcodes = ['3xKsf9qdS1CyvXSMEid6g8', '7BaDRi8gmRnOducT6KaOfc', '7uJH0YdSZL6psjxI6Xy08b', '5dpg0w2gqdPMwAPrZ0zj1K', '1p5B1jJk75RAblZCTnzUTU']
WarmPrecip_trackcodes = ['2X1v0cUbDZXzi21L0yAH79', '3VeuvRWfJLSOzeVPrgi5MR', '6gpcs5eMhJwax4mIfKDYQk', '71lyUtIPCMAYjTwwwxy2jW', '1F2HXIJrE2Xn81OzT2zZvW']
MildDry_trackcodes = ['2g2GkH3vZHk4lWzBjgQ6nY', '2Nh2cMryoXl7BrZoIeN2Pr', '5mCPDVBb16L4XQwDdbRUpz', '33pDmPx2pkJ7JCAQpRDQ9B', '4h90qkbnW1Qq6pBhoPvwko']
MildPrecip_trackcodes = ['371H6HjS4SXGbQ9IVfFUIL', '3FtYbEfBqAlGO46NUDQSAt', '40jxatV29fk82SAziDsPqN', '4ntsD33Pgsgk8rUyZLLiUV', '2c2csx4OTYtbkzvbSTXlGY']
ColdDry_trackcodes = ['6n3HGiq4v35D6eFOSwqYuo', '7D6RDU4YoT60JyK40QyfjT', '3S1Q4qmbzVZWwMOtZ0RNcs', '0WgBb7XgdtbUW0GlYel9mH', '3XVozq1aeqsJwpXrEZrDJ9']
ColdPrecip_trackcodes = ['7bJ4mu7MHa3rHiNyKjOoSl', '005qPpqZ4PYk9JTh2EijJr', '0utlOiJy2weVl9WTkcEWHy', '5MNLzsJJNcpp01Sav4xfvX', '6mFkJmJqdDVQ1REhVfGgd1']

# loop to get user input
while True:
    try:
        print("Please enter a state and a city in the following format:\n")
        print("NY")
        print("New York\n")
        loc_state = input("Enter state: ")
        loc_city = input("Enter city: ")
        conditions = get_wunderground_weather(loc_city, loc_state)
        break # breaks if state and city are valid
    except KeyError:
        print("Please enter a valid state and city in the correct format.")
        print("For example: NY -> Niagara Falls\n")

# Cold and Dry
if conditions['tempF'] < 32 and conditions['precip_today'] < 0.05:
    print("Playlist: Ice Cold")
    get_track_name(ColdDry_trackcodes)
    time.sleep(6)
    webbrowser.open_new(ColdDry_trackcodes)
# Cold and Precip
elif conditions['tempF'] < 32 and conditions['precip_today'] >= 0.05:
    print("Playlist: Cold & Snowy")
    get_track_name(ColdPrecip_trackcodes)
    time.sleep(6)
    webbrowser.open_new(ColdPrecip_url)
# Mild and Dry
elif conditions['tempF'] < 60 and conditions['tempF'] >= 32 and conditions['precip_today'] < 0.05:
    print("Playlist: Coolin' It")
    get_track_name(MildDry_trackcodes)
    time.sleep(6)
    webbrowser.open_new(MildDry_url)
# Mild and Precip
elif conditions['tempF'] < 60 and conditions['tempF'] >= 32 and conditions['precip_today'] >= 0.05:
    print("Playlist: Mild Weather w/ Rain")
    get_track_name(MildPrecip_trackcodes)
    time.sleep(6)
    webbrowser.open_new(MildPrecip_url)
# Warm and Dry
elif conditions['tempF'] >= 60 and conditions['precip_today'] < 0.05:
    print("Hot Daze")
    get_track_name(WarmDry_trackcodes)
    time.sleep(6)
    webbrowser.open_new(WarmDry_url)
# Warm and Wet
elif conditions['tempF'] >= 60 and conditions['precip_today'] >= 0.05:
    print("Hot&Rainy")
    get_track_name(WarmPrecip_trackcodes)
    time.sleep(6)
    webbrowser.open_new(WarmPrecip_url)
```
